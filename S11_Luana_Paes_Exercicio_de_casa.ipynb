```python
import numpy as np
```

```python
temperatura = np.genfromtxt('temperatura_RJ.csv', delimiter=',', skip_header=1)
```

```python
data = np.transpose(temperatura)
```

```python
print(temperatura)
```

```python
temperatura_hora = data[1]
print(temperatura_hora)
```

```python
 #Estatística Descritiva
#média
np.mean(temperatura_hora)
```

```python
#mediana
np.median(temperatura_hora)
```

```python
#desvio padrão
np.std(temperatura_hora)
```

```python
#quatis
Min = np.min(temperatura_hora)
Q1 = np.quantile(temperatura_hora,0.25)
Q2 = np.quantile(temperatura_hora,0.5)
Q3 = np.quantile(temperatura_hora,0.75)
Max = np.max(temperatura_hora)
print("valor minino:", Min)
print("valor 25%:", Q1)
print("valor 50%:", Q2)
print("valor 75%:", Q3)
print("valor maximo:", Max)
```

```python
#gráfico de boxplot
import matplotlib.pyplot as plt
```

```python
plt.boxplot(temperatura_hora)
```

##Com base em dados que vc coletou monte um gráfico (linhas, barra ou dispersão) que melhor represente os dados

```python
data = np.genfromtxt('Dados Olinpiada - Medalhas Brasil.csv', delimiter=',', skip_header = 1)
```

```python
data = np.transpose(data)
```

```python
ano = data[0]
medalhas_ouro = data[1]
medalhas_prata = data[2]
medalhas_bronze = data[3]
medalhas_total = data[4]
```

```python
plt.grid()
plt.plot(ano,medalhas_ouro, color = '#bca111', label='Ouro')
plt.plot(ano,medalhas_prata, color = '#a5a5a1', label='Prata')
plt.plot(ano,medalhas_bronze, color = '#B8860B', label='Bronze')
plt.plot(ano,medalhas_total, color = '#000000', label='Total')
plt.legend()
```

```python
plt.grid()
plt.scatter(ano,medalhas_ouro, color = '#bca111', label='Ouro')
plt.scatter(ano,medalhas_prata, color = '#a5a5a1', label='Prata')
plt.scatter(ano,medalhas_bronze, color = '#B8860B', label='Bronze')
plt.scatter(ano,medalhas_total, color = '#000000', label='Total')
plt.legend()
```
